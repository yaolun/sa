### Fucntion for correcting source velocity

```python
def vlsr_correction(freq, vlsr):
    """
    freq: the frequency array of spectrum
    vlsr: the vlsr of source in the unit of km/s
    """
    
    c = 29979245800.0
    return freq + freq*(vlsr*1e5)/c
```

### Literatures on the v$_{lsr}$ of BHR 71

Yildiz+2013: -4.4 km/s (C$^{18}$O or C$^{17}$O)  
Chen+2008: -4.35+/-0.02 km/s for BHR 71 E; -4.42+/-0.02 km/s for BHR 71 W. (N$_{2}$H$^{+}$)  
Parise+2006: -4.5 km/s (I think they just use Bourke+1997 value)  
Henning & Launhardt 1998: -4.4 km/s (CS)
Bourke+1997: -4.63 km/s (all C$^{18}$O spectra); -4.45 km/s (all NH$_{3}$ spectra)


### Correct the v$_{lsr}$ for frequency

```python
datadir = '/Users/yaolun/GoogleDrive/research/bhr71_infall/analysis/'
vers = [('imcontsub', '.imcontsub.'), ('pipeline', '.pipeline.'), ('reduced_spectra', '_')]
spws = ['0', '1', '2', '3']
vlsr = -4.45

for ver in vers:
    for spw in spws:
        spec_dum = ascii.read(datadir+ver[0]+'/spw'+spw+ver[1]+'scriptExtraction.txt')
        spec = pd.DataFrame(spec_dum.as_array())
        spec.columns = spec_dum.meta['comments'][-1].split()
        spec['Freq_corrected'] = vlsr_correction(spec['frequency_(MHz)']/1e3, vlsr)
        
        foo = open(datadir+ver[0]+'/spw'+spw+ver[1]+'scriptExtraction.vlsr_corrected.txt', 'w')
        foo.write('{:<16s}  {:<16s}\n'.format('Frequency(GHz)', 'Flux_density(Jy)'))
        for i, (f, jy) in enumerate(zip(spec['Freq_corrected'], spec['Flux_density_(Jy)'])):
            foo.write('{:<16.8f}  {:<16.8e}\n'.format(f, jy))
        foo.close()
```

## Plot the pipeline data extracted with the same region as manually imaging products

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import astropy.constants as const
c = const.c.cgs.value

spws = ['0', '1', '2', '3']
datadir = '/Users/yaolun/GoogleDrive/research/bhr71_infall/analysis/pipeline/'

linelist = ['HCN 4-3', 'HCO$^{+}$ 4-3', 'CS 7-6', 'H$^{13}$CN 4-3']
restfreqs = [354.505473, 356.734242, 342.882857, 345.339756]

# the size of extracted region
region = (0.5213, 0.394)

def Sjy2t(freq, beam=(0.4, 0.28)):
    s_maj, s_min = beam
    return 13.6*(300/freq)**2/s_maj/s_min

for i, spw in enumerate(spws):
    spec_dum = ascii.read(datadir+'spw'+spw+'.pipeline.scriptExtraction.txt')
    spec = pd.DataFrame(spec_dum.as_array())
    spec.columns = spec_dum.meta['comments'][-1].split()
    
    fig = plt.figure(figsize=(24,4))
    ax = fig.add_subplot(111)
    
    ax.plot(spec['Velocity_(km/s)'][:-1], spec['Flux_density_(Jy)'][:-1]*Sjy2t(restfreqs[i], beam=region),
            color='b', label=linelist[i])
    # plot zero-temperature baseline
    ax.axhline(0, color='Gray', linewidth=0.7)
    
    # source velocity
    ax.axvline(-4.45, color='k', linestyle='-', linewidth=0.7)

    ax.legend(loc='best', fontsize=16)
    # ax.set_xlim([-9.4,0.4])
    ax.set_xlabel(r'velocity [km s$^{-1}$]', fontsize=20)
    ax.set_ylabel(r'$T_{\rm mb}$ [K]', fontsize=20)
    ax.tick_params('both',labelsize=16,width=1.5,which='major',pad=5,length=5)
    ax.tick_params('both',labelsize=16,width=1.5,which='minor',pad=5,length=2.5)

    [ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
    ax.minorticks_on()

    fig.savefig('/Users/yaolun/GoogleDrive/research/bhr71_infall/analysis/pipeline/spw'+spw+'.pipeline.scriptExtraction.pdf',
                format='pdf', dpi=300, bbox_inches='tight')
```

## Comparision between the pipeline products and manually imaged products

```python
import astropy.constants as const
c = const.c.cgs.value

spws = ['0', '1', '2', '3']
datadir = '/Users/yaolun/GoogleDrive/research/bhr71_infall/analysis/'

linelist = ['HCN 4-3', 'HCO$^{+}$ 4-3', 'CS 7-6', 'H$^{13}$CN 4-3']
restfreqs = [354.505473, 356.734242, 342.882857, 345.339756]

# the size of extracted region
region = (0.5213, 0.394)

# not sure this is correct
def Sjy2t(freq, beam=(0.4, 0.28)):
    s_maj, s_min = beam
    return 13.6*(300/freq)**2/s_maj/s_min

for i, spw in enumerate(spws):
    spec_dum = ascii.read(datadir+'/imcontsub/spw'+spw+'.imcontsub.scriptExtraction.txt')
    spec = pd.DataFrame(spec_dum.as_array())
    spec.columns = spec_dum.meta['comments'][-1].split()
    
    pipe_spec_dum = ascii.read(datadir+'/pipeline/spw'+spw+'.pipeline.scriptExtraction.txt')
    pipe_spec = pd.DataFrame(pipe_spec_dum.as_array())
    pipe_spec.columns = pipe_spec_dum.meta['comments'][-1].split()
    
    fig = plt.figure(figsize=(24,4))
    ax = fig.add_subplot(111)
    
    ax.plot(spec['Velocity_(km/s)'][:-1], spec['Flux_density_(Jy)'][:-1]*Sjy2t(restfreqs[i], beam=region),
            color='b', label=linelist[i])
    # pipeline
    ax.plot(pipe_spec['Velocity_(km/s)'][:-1], pipe_spec['Flux_density_(Jy)'][:-1]*Sjy2t(restfreqs[i], beam=region),
            color='r', label=linelist[i]+' pipeline')
    
    # source velocity
    ax.axvline(-4.45, color='k', linestyle='-', linewidth=0.7)

    ax.legend(loc='best', fontsize=16)
    # ax.set_xlim([-9.4,0.4])
    ax.set_xlabel(r'velocity [km s$^{-1}$]', fontsize=20)
    ax.set_ylabel(r'$T_{\rm mb}$ [K]', fontsize=20)
    ax.tick_params('both',labelsize=16,width=1.5,which='major',pad=5,length=5)
    ax.tick_params('both',labelsize=16,width=1.5,which='minor',pad=5,length=2.5)

    [ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
    ax.minorticks_on()

#     fig.savefig('/Users/yaolun/GoogleDrive/research/bhr71_infall/analysis/spw'+spw+'.imcontsub.scriptExtraction.pdf',
#                 format='pdf', dpi=300, bbox_inches='tight')
```
