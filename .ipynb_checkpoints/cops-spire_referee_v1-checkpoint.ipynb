Function for calculating $\beta$ at PACS bands

```python
def spectral_index(spec, waveBands, freqRange, plot=False):

    # Note that the spectral index works in frequency
    import matplotlib.pyplot as plt
    from astropy.modeling import models, fitting, powerlaws, custom_model, blackbody
    from scipy.interpolate import interp1d
    import astropy.constants as const
    import numpy as np
    c = const.c.cgs.value

    # filter NaN values
    spec = spec[np.isnan(spec['Flux_Density(Jy)']) == False]
    
    if plot:
        # initiate figure
        fig = plt.figure()
        ax = fig.add_subplot(111)
        
        ax.plot(c/1e5/spec['Wavelength(um)'], spec['Flux_Density(Jy)'], color='b')
    
    f = interp1d(c*1e4/spec['Wavelength(um)'], spec['Flux_Density(Jy)'])
    fitted_alpha = []
    fitted_alpha_err = []

    for band in waveBands:
        # frequency of the representing wavelength
        x_ref = c/band*1e4
        # get the flux at the representing frequency
        amp = f(x_ref)
        # select a range of spectrum for fitting the power-law by the given range
        freq_dum = (c*1e4/spec['Wavelength(um)'])\
                   [(c*1e4/spec['Wavelength(um)'] >= x_ref-freqRange*1e9) & (c*1e4/spec['Wavelength(um)'] <= x_ref+freqRange*1e9)]
        flux_dum = spec['Flux_Density(Jy)']\
                   [(c*1e4/spec['Wavelength(um)'] >= x_ref-freqRange*1e9) & (c*1e4/spec['Wavelength(um)'] <= x_ref+freqRange*1e9)]

        fitter = fitting.LevMarLSQFitter()
            
        # setup the power-law model
        pow_model = powerlaws.PowerLaw1D(amp, x_ref, 0)
        fit = fitter(pow_model, freq_dum, flux_dum)
        
        # take negative sign because the frequency array is reversed
        fitted_alpha.append(-fit.alpha.value)
        
        if plot:
            ax.plot(freq_dum/1e9, fit(freq_dum), '-', color='k',
                    label='{:<4.2f}'.format(-fit.alpha.value))
            
    if plot:
        [ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
        ax.minorticks_on()
        ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=10,length=5)
        ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=10,length=2.5)
        ax.set_xlabel('Frequency [GHz]', fontsize=18)
        ax.set_ylabel('Flux Density [Jy]', fontsize=18)
        return (fitted_alpha), fig
    else:
        return (fitted_alpha)
```

Setup source list

```python
obj_list = ['RCrA-IRS7B','RCrA-IRS7C','HH46','L723-MM','L1014',
            'L1157','Ced110','BHR71','IRAS03245','L1551-IRS5',
            'L1455-IRS3','B1-a','B1-c','IRAS03301','TMR1',
            'TMC1A','TMC1','IRAS15398','RNO91','GSS30-IRS1',
            'VLA1623','WL12','RCrA-IRS5A','L483','B335',
            'DKCha']
archive_dir = '/Volumes/SD-Mac/CDF_archive_v2/'
```

For each source:  
1. Load the PACS spectrum.  
2. Measure $\beta$ at 70, 100, and 160 $\mu$m.  
3. Load the photometry.  
4. Apply the color correction.  

```python
def ccLookup(CC_table, beta, name, obj):
    from scipy.interpolate import interp1d
    f = interp1d(CC_table['beta'], CC_table[name])
    
    if beta >= 3:
        cc = CC_table[name][CC_table['beta'] == 3].data
        print(obj+' '+name.split('CC_')[1]+' beta > 3 ('+str(beta)+')')
    elif beta <= -3:
        cc = CC_table[name][CC_table['beta'] == -3].data
        print(obj+' '+name.split('CC_')[1]+' beta < -3 ('+str(beta)+')')
    else:
        cc = f(beta)
    
    return float(cc)

import numpy as np

bands = ['CC_70', 'CC_100', 'CC_160']

for o in obj_list:
    try:
        spec = ascii.read(archive_dir+o+'/pacs/advanced_products/'+o+'_pacs_weighted_continuum.txt')
    except:
        continue
    beta = spectral_index(spec, [70, 100, 160], 200, plot=False)
    # get color correction factors
    cc_list = [ccLookup(CC_table, b, n, o) for i, (b, n) in enumerate(zip(beta, bands))]
    
    # load photometry
    phot = ascii.read(archive_dir+o+'/pacs/data/'+o+'_pacs_phot.txt',
                      data_start=4, names=['Wavelength(um)', 'Flux_Density(Jy)', 'Uncertainty(Jy)'])
    # take average of measurements at the same wavelength
    phot_wave = np.sort(list(set(phot['Wavelength(um)'])))
    phot_flux = [np.mean(phot['Flux_Density(Jy)'][phot['Wavelength(um)'] == w]) for w in phot_wave]
    phot_unc = [np.sum(phot['Uncertainty(Jy)'][phot['Wavelength(um)'] == w]**2)**0.5 for w in phot_wave]
    
    phot_flux_cc = [f/cc for i, (f, cc) in enumerate(zip(phot_flux, cc_list))]
    
    # write out the color-corrected photometry
    foo = open(archive_dir+o+'/pacs/data/'+o+'_pacs_phot_cc.txt', 'w')
    foo.write('{:<18s}  {:<18s}\n'.format('Wavelength(um)', 'Flux_Density(Jy)'))
    for i, (w, f) in enumerate(zip(phot_wave, phot_flux_cc)):
        foo.write('{:<18.1f}  {:<18.8f}\n'.format(w, f))
    foo.close()
```

The uncertainty of indice of power-law is not constrained, and sometime the index may exceed the limit of the color correction table.  
Tried to use modified blackbody fitting, but the problem remains.

```python
%matplotlib inline
import astropy.constants as const
import matplotlib.pyplot as plt
c = const.c.cgs.value
o = 'BHR71'
spec = ascii.read(archive_dir+o+'/pacs/advanced_products/'+o+'_pacs_weighted_continuum.txt')
noise = ascii.read(archive_dir+o+'/pacs/advanced_products/'+o+'_pacs_weighted_residual_spectrum.txt')
plt.plot(c/1e5/spec['Wavelength(um)'], spec['Flux_Density(Jy)']+noise['Flux_Density(Jy)'])

from astropy.modeling import blackbody

def mod_BB(x, beta=1, temp=35., norm=1e-12, norm2=5e13):
    tau = norm*x**beta
    modifier = (1-np.exp(-tau))*norm2
    return (modifier * blackbody.blackbody_nu(x, temp).value)

plt.plot(c/1e5/spec['Wavelength(um)'], mod_BB(c/spec['Wavelength(um)']*1e4))
```

```python
def fit_modBB(o, archive_dir, cc_table='/Users/yaolun/programs/misc/hipe_script/colorCorrectionPACS_modBB.txt'):

    # Note that the spectral index works in frequency
    import matplotlib.pyplot as plt
    from astropy.modeling import models, fitting, powerlaws, custom_model, blackbody
    from scipy.interpolate import interp1d
    import astropy.constants as const
    from astropy.io import ascii
    import numpy as np
    c = const.c.cgs.value

    CC_table = ascii.read(cc_table)
    f_cc = interp1d(CC_table['temp'], CC_table['CC_160'])
    
    spec = ascii.read(archive_dir+o+'/pacs/advanced_products/'+o+'_pacs_weighted_continuum.txt')
    noise = ascii.read(archive_dir+o+'/pacs/advanced_products/'+o+'_pacs_weighted_residual_spectrum.txt')

    # filter NaN values
    spec = spec[np.isnan(spec['Flux_Density(Jy)']) == False]
    noise = noise[np.isnan(spec['Flux_Density(Jy)']) == False]
    weight = 1/(spec['Flux_Density(Jy)']*0.1)

    # initiate figure
    fig = plt.figure()
    ax = fig.add_subplot(111)

    p, = ax.plot(c/1e5/spec['Wavelength(um)'], spec['Flux_Density(Jy)'], color='b')

    f = interp1d(c*1e4/spec['Wavelength(um)'], spec['Flux_Density(Jy)'])
    fitted_alpha = []
    fitted_alpha_err = []

    # frequency of the representing wavelength
    x_ref = c/160*1e4
    # get the flux at the representing frequency
    amp = f(x_ref)
    # select a range of spectrum for fitting the power-law by the given range
    selector = (c*1e4/spec['Wavelength(um)'] <= c/110*1e4)
    # selector = np.ones_like(spec['Wavelength(um)'], dtype='bool')

    freq_dum = (c*1e4/spec['Wavelength(um)'][selector])
    flux_dum = spec['Flux_Density(Jy)'][selector]

    fitter = fitting.LevMarLSQFitter()

    # setup modified-BB model
    @custom_model
    def mod_BB(x, beta=2, temp=20., norm=amp/blackbody.blackbody_nu(x_ref, 20).value/x_ref**2):
        modifier = norm*x**beta
        return (modifier * blackbody.blackbody_nu(x, temp).value)

    fit = fitter(mod_BB(), freq_dum, flux_dum, weights=weight[selector])

    print(o, fit.parameters, fitter.fit_info['message'], fitter.fit_info['ierr'])

    ax.plot(freq_dum/1e9, fit(freq_dum), '-', color='k')
    #                     label='{:<4.2f}+/-{:<4.2f}'.format(-fit.alpha.value, fitter.fit_info['param_cov'][2,2]**0.5))

    [ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
    ax.minorticks_on()
    ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=10,length=5)
    ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=10,length=2.5)
    ax.set_xlabel('Frequency [GHz]', fontsize=18)
    ax.set_ylabel('Flux Density [Jy]', fontsize=18)
#     ax.text(0.7, 0.8, o, transform=ax.transAxes, fontsize=18)
    ax.legend([p], [o+r', $\beta$={:<.1f}, T={:<.1f}'.format(fit.parameters[0], fit.parameters[1])], loc='best', fontsize=16)
    
    return fit.parameters, f_cc(fit.parameters[1])
```

```python
fitted_params = []
for o in obj_list:
    if o in ['HH46', 'L723-MM', 'Ced110', 'IRAS15398', 'RNO91', 'L483']:
        continue
    params = fit_modBB(o, archive_dir)
    fitted_params.append(params)
```
