```python
# first to run everytime
%config InlineBackend.figure_format = 'retina'
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import os
from astropy.io import ascii
import pandas as pd
```

### Fucntion for correcting source velocity

```python
def vlsr_correction(freq, vlsr):
    """
    freq: the frequency array of spectrum
    vlsr: the vlsr of source in the unit of km/s
    """
    
    c = 29979245800.0
    return freq + freq*(vlsr*1e5)/c
```

### Literatures on the v$_{lsr}$ of BHR 71

Yildiz+2013: -4.4 km/s (C$^{18}$O or C$^{17}$O)  
Chen+2008: -4.35+/-0.02 km/s for BHR 71 E; -4.42+/-0.02 km/s for BHR 71 W. (N$_{2}$H$^{+}$)  
Parise+2006: -4.5 km/s (I think they just use Bourke+1997 value)  
Henning & Launhardt 1998: -4.4 km/s (CS)
Bourke+1997: -4.63 km/s (all C$^{18}$O spectra); -4.45 km/s (all NH$_{3}$ spectra)


### Work flow for processing ALMA spectra extracted with $\texttt{specflux}$

1. Read the spectra after performed $\texttt{imcontsub}$
1. Correct for the source velocity.
2. re-fit the baseline.
3. Output(s)
    - ASCII files
    - FITS for CLASS

```python
import astropy.constants as const
c = const.c.cgs.value

spws = ['0', '1', '2', '3']
datadir = '/Users/yaolun/GoogleDrive/research/bhr71_infall/analysis/imcontsub/'

linelist = ['HCN 4-3', 'HCO$^{+}$ 4-3', 'CS 7-6', 'H$^{13}$CN 4-3']
restfreqs = [354.505473, 356.734242, 342.882857, 345.339756]

# source velovity in km/s
vlsr = -4.45

# the size of extracted region
region = (0.5213, 0.394)

# baseline region for each spw
base_regions = [[354.415, 354.44, 354.58, 354.595],
                [356.635, 356.655, 356.67, 356.695, 356.81, 356.83],
                [342.80, 342.86, 342.90, 342.92, 342.955, 342.965],
                [345.19, 345.20, 345.26, 345.28, 345.36, 345.39]]

# not sure this is correct
def Sjy2t(freq, beam=(0.39, 0.27)):
    s_maj, s_min = beam
    return 13.6*(300/freq)**2/s_maj/s_min

for i, spw in enumerate(spws):

    # read in the spectra
    spec_dum = ascii.read(datadir+'spw'+spw+'.imcontsub.scriptExtraction.txt')
    spec = pd.DataFrame(spec_dum.as_array())
    spec.columns = spec_dum.meta['comments'][-1].split()
    
    # apply correction for source velocity
    spec['Freq_corrected'] = vlsr_correction(spec['frequency_(MHz)']/1e3, vlsr)
    
    # re-fit the baseline for each spw
    # select the baseline for fitting
    base_select = np.zeros_like(spec['Freq_corrected'], dtype='bool')
    for j in range(len(base_regions[i])//2):
        select = (spec['Freq_corrected'] >= base_regions[i][2*j]) & (spec['Freq_corrected'] <= base_regions[i][2*j+1])
        base_select = np.logical_or(base_select, select)
        
    # use 1st order polynomial first
    p, cov = np.polyfit(spec['Freq_corrected'][base_select], spec['Flux_density_(Jy)'][base_select], 1, cov=True)
    
    # remove the baseline
    spec['baseline'] = p[0]*spec['Freq_corrected']+p[1]
    spec['Flux_corrected'] = spec['Flux_density_(Jy)']-spec['baseline']
    
    # write the processed spectra to files
    spec.to_csv(datadir+'spw'+spw+'.imcontsub.scriptExtraction.processed.txt', index=False)
    
    # plot
    fig = plt.figure(figsize=(12,4))
    ax = fig.add_subplot(111)
    
    ax.plot(spec['Freq_corrected'], spec['baseline']*Sjy2t(restfreqs[i], beam=region), color='g', label='baseline')
    ax.plot(spec['Freq_corrected'], spec['Flux_corrected']*Sjy2t(restfreqs[i], beam=region),
            color='Gray', alpha=0.7, linewidth=0.7, label='baseline subtracted')
    
    ax.plot(spec['Freq_corrected'][:-1], spec['Flux_density_(Jy)'][:-1]*Sjy2t(restfreqs[i], beam=region),
            color='b', label=linelist[i], linewidth=0.7)
    
    # plot the regions for baseline fitting
    for j in range(len(base_regions[i])//2):
        ax.axvspan(base_regions[i][2*j], base_regions[i][2*j+1], alpha=0.3, color='r')
    
    # plot the zero baseline
    ax.axhline(0, color='r', linewidth=1.0)

    ax.set_ylim([-5, 10])
    ax.legend(loc='best', fontsize=12)
    ax.set_xlabel(r'frequency [GHz]', fontsize=20)
    ax.set_ylabel(r'$T_{\rm mb}$ [K]', fontsize=20)
    ax.tick_params('both',labelsize=16,width=1.5,which='major',pad=5,length=5)
    ax.tick_params('both',labelsize=16,width=1.5,which='minor',pad=5,length=2.5)

    [ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
    ax.minorticks_on()

#     fig.savefig('/Users/yaolun/GoogleDrive/research/bhr71_infall/analysis/imcontsub/spw'+spw+'.imcontsub.scriptExtraction.pdf',
#                 format='pdf', dpi=300, bbox_inches='tight')
```
