```python
# test if the extracted spectrum with "specflux" with a region file works properly
from astropy.io import ascii
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
# region area
region_area = np.pi*0.25975*0.1914   # arcsec2
beam_area = 0.130253814368           # aresec2

spw_label = [r'HCN $J=4\rightarrow3$', r'HCO$^{+}\,J=4\rightarrow3$',
             r'CS $J=7\rightarrow6$', r'H$^{13}$CN $J=4\rightarrow3$']

for i, spw in enumerate(['25', '27', '29', '31']):
    filename = '/Volumes/SD-Mac/alma_bhr71/spw'+spw+'_spectrum.txt'
    data = ascii.read(filename, names=['Channel','number_of_unmasked_pixels',
                                       'frequency_(MHz)','Velocity_(km/s)','Flux_density_(Jy)'])

    fig = plt.figure(figsize=(18,4))
    ax = fig.add_subplot(111)
    spec, = ax.plot(data['Velocity_(km/s)'], data['Flux_density_(Jy)']/(region_area/beam_area),
                    color='b', linewidth=0.7)
    ax.legend([spec], [spw_label[i]], fontsize=16, loc='upper right')
    ax.set_xlabel(r'Velocity [km s$^{-1}$]', fontsize=18)
    ax.set_ylabel(r'Flux Density [Jy beam$^{-1}$]', fontsize=18)
    ax.minorticks_on()
    [ax.spines[axis].set_linewidth(1) for axis in ['top','bottom','left','right']]
    ax.tick_params('both',labelsize=14,width=1.5,which='major',pad=15,length=5)
    ax.tick_params('both',labelsize=14,width=1.5,which='minor',pad=15,length=2.5)
    fig.savefig('/Volumes/SD-Mac/alma_bhr71/spw'+spw+'_spectrum.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
